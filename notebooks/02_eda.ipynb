```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("../data/SaberPro.xlsx")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
col_puntaje = [col for col in df.columns if 'PUNT' in col.upper()][0]
print("Columna de puntaje:", col_puntaje)
```

```python
df[col_puntaje].hist()
plt.title("Distribución Puntaje Global")
plt.show()
```

```python
sns.boxplot(x=df[col_puntaje])
plt.title("Outliers Puntaje")
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr)
plt.title("Mapa de calor")
plt.show()
```

```python
col_estrato = [col for col in df.columns if 'ESTRATO' in col.upper()]

if col_estrato:
    sns.boxplot(x=df[col_estrato[0]], y=df[col_puntaje])
    plt.title("Estrato vs Puntaje")
    plt.show()
```
